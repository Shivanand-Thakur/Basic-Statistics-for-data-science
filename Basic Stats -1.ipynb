```python
# Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "/content/drive/MyDrive/dataset/sales_data_with_discounts.csv"
df = pd.read_csv(file_path)

print("Dataset Info:\n")
print(df.info())

# Select numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Descriptive statistics: mean, median, std
descriptive_stats = df[num_cols].agg(['mean', 'median', 'std']).transpose()

# Mode calculation
modes = df[num_cols].mode().iloc[0]
descriptive_stats['mode'] = modes

print("\nDescriptive Statistics for Numerical Columns:\n")
print(descriptive_stats)

# =========================
# Data Visualization
# =========================

# Histograms
for col in num_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Boxplots
for col in num_cols:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col], vert=True, patch_artist=True)
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Bar Charts for categorical variables
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    plt.figure(figsize=(6,4))
    df[col].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# =========================
# Conclusion (automatic summary)
# =========================
print("\nConclusion:")
print("- Numerical columns show variation in sales, discounts, and pricing.")
print("- Histograms reveal skewness in sales and price distributions.")
print("- Boxplots highlight outliers in Total Sales Value and Discount Amount.")
print("- Categorical analysis shows dominant brands, models, and cities.")
```
